```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
month = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
totalFare = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create plot
plt.plot(month,totalFare)
```

```python
y_axis_larger = []
for data in totalFare:
  y_axis_larger.append(data*3) #increases the size of all the points on the plot
# ^Can do the same thing by using:
    # new_list = [expression for item in list if conditional]

plt.scatter(month, totalFare, s=y_axis_larger)
```

```python
# Using Object-Orientated Interface
fig, ax = plt.subplots()

# credit goes to: https://stackabuse.com/matplotlib-change-scatter-plot-marker-size/
s = [5*s for s in totalFare] #this multiplies the totalFare array by 5
# vs totalFare*5 which just prints totalFare array 5x

ax.scatter(totalFare, month, s = s, color = "skyblue", alpha = 0.8, linewidths=2, edgecolors="k", label="Boston")
#color credit goes to: https://matplotlib.org/stable/gallery/color/named_colors.html
#transparency credit: https://stackoverflow.com/questions/30108372/how-to-make-matplotlib-scatterplots-transparent-as-a-group
#documentation: https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.scatter.html

ax.set_title("Pyber Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Month")
ax.legend()
ax.set_ylim("Dec","Jan") #bottom to top / left to right (x axis)
#still want to figure out how to adjust the graph to be able to see the full 
    #points that lie within Jan and Dec
```

```python
#matlab matplotlib
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))

plt.pie(totalFare, labels=month, explode=explode_values, autopct='%.1f%%', colors=colors)
plt.show()
```

```python
#object orientated interface
explode_values = (0, 0, 0.1, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots(figsize=(8, 8))

#Add: percentage, fig size 8x8, explode two highest months, 
    # add a shadow, add a start angle so Jan is at the top, 
    # reverse the month order in clockwise direction
    # Add new colors

ax.pie(totalFare, labels=month, explode=explode_values, autopct='%1.1f%%', 
      shadow=True, startangle=96, counterclock=False, colors=colors)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(month, totalFare)
ax.set_xticks(np.arange(0, 51, step=5.0))
ax.set_ylim("Dec","Jan")
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(month, totalFare)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
