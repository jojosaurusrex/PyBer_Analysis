```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
month = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
totalFare = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create plot
plt.plot(month,totalFare)
```

```python
#MATLAB graphing plots
#plt.plot(month,totalFare)

#Object-Oriented Interface
#fig, ax = plt.subplots()
#zx.plot(month,totalFare)

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(month,totalFare)
#fig references the figure
#ax references the axes of the figure and allows them to be annotated
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(month,totalFare)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(month,totalFare)

#different ways to do the same thing
```

```python
#create the plot and add a label for the legend
plt.plot(month, totalFare, label='Boston')

#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45
plt.ylim(0, 45)

#Create a title
plt.title("Pyber Fare by Month")

#Add the legend
plt.legend()
```

```python
#Create the plot
plt.plot(month, totalFare, marker="*", color="blue", linewidth=2, label='Boston')

#Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45
plt.ylim(0, 45)

#Create a title
plt.title("Pyber Fare by Month")

#Add the legend
plt.legend()

#Add a grid
plt.grid()
```

```python
#Create the plot
plt.plot(month, totalFare, marker="d", color="green", linewidth=2, label='Boston')

#Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit
plt.ylim(0, 50)

#Create a title
plt.title("Pyber Fare by Month")

#Add the legend
plt.legend()

#Add a grid
plt.grid()
```

```python
#Create plot
plt.bar(month, totalFare, color="green", label="Boston")

#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Create a title
plt.title("Pyber Fare by Month")

#Add legend
plt.legend()
```

```python
#Create plot / dont't grab two plots on the same output
plt.barh(month, totalFare)
plt.gca().invert_yaxis()
```

```python
plt.barh(month, totalFare, color="magenta", label="Boston")
plt.gca().invert_yaxis()

plt.title("Pyber Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Month")
plt.legend()
```

```python
#Vertical bar chart with ax.plt()
fig, ax = plt.subplots()
ax.bar(month, totalFare, color="k", label="New York") #k = black
```

```python
fig, ax = plt.subplots()
ax.barh(month, totalFare, color="cyan", label="Chicago")

ax.set_title("Pyber Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Month")
ax.legend()
ax.set_ylim("Dec", "Jan") #bottom to top / left to right (x axis)
```
